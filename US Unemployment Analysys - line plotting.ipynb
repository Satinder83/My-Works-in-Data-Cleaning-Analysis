# Line Charts Plotting using Matplotlib

The data is based on  United States Bureau of Labor Statistics (BLS) data from monthly unemployment rate as a CSV from January 1948 to August 2016

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

unrate = pd.read_csv('unrate.csv')

unrate.head()
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
unrate.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Line plot for first 12 months data for year 1948

```python
first_12 = unrate.head(12)
plt.plot(first_12.DATE, first_12.VALUE, color = 'orange', linewidth= 2)
plt.xticks(rotation= 45)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.legend()
plt.show()
```

#### plotting data for next 4 years in different subplots

```python
fig = plt.figure(figsize=(12,12))

for i in range(5):
    ax= plt.subplot(5, 1, i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])
```

```python
unrate['Month'] = unrate['DATE'].dt.month
```

##### The same line diagramme can be plotted in a single plot to derive more meaningful analysis for 4 years

```python
fig = plt.figure(figsize = (10, 6))
colour = ['red', 'blue','green','orange','black',]

for i in range(5):
    subset_start = i * 12
    subset_end = (i+1) * 12
    subset = unrate[subset_start:subset_end]
    plt.plot(subset['Month'], subset['VALUE'], c = colour[i], label = '{}'.format(1948+i))
        
plt.legend(loc = 'upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel('Month, Integer')
plt.ylabel('Unemployment Rate, Percent')
plt.show()
```
